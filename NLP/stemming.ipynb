```python
## Classification Problem
## Comments of product is a positive review or negative review
## Reviews----> eating, eat,eaten [going,gone,goes]--->go

words=["eating","eats","eaten","writing","writes","programming","programs","history","finally","finalized"]
```

```python
from nltk.stem import PorterStemmer
```

```python
stemming=PorterStemmer()
```

```python
for word in words:
    print(word+"---->"+stemming.stem(word))
```

```python
stemming.stem("funny")
```

```python
from nltk.stem import RegexpStemmer
```

regexp (str or regexp) – The regular expression that should be used to identify morphological affixes.

min (int) – The minimum length of string to stem

```python
reg_stemmer=RegexpStemmer('ing$|s$|e$|able$', min=4)
```

```python
reg_stemmer.stem('eating')
```

```python
reg_stemmer.stem('stemming')
```

```python
from nltk.stem import SnowballStemmer
```

```python
snowballsstemmer=SnowballStemmer('english') #select language for looking up the stem of the word
```

```python
for word in words:
    print(word+"---->"+snowballsstemmer.stem(word))
```

```python
stemming.stem("fairly"),stemming.stem("sportingly")
```

```python
snowballsstemmer.stem("fairly"),snowballsstemmer.stem("sportingly")
```
